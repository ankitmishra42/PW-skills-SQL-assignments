Q1. Create a New Database and Table for Employees
 
Task: Create a new database named company_db and Create a table named  
employee with the following columns.

```python
"""
    -- Creating a database
    CREATE DATABASE company_db;
    USE company_db;
"""
"""
    -- Creating Table
    CREATE TABLE employee (
      employee_id INT PRIMARY KEY,
      first_name VARCHAR(50),
      last_name VARCHAR(50),
      department VARCHAR(50),
      salary INT,
      hire_date DATE
    );
"""
```

Q3. Display All Employee Records Sorted by Salary (Lowest to Highest).

```python
"""
    SELECT * FROM employee 
    ORDER BY salary;
"""
```

Q4. Show Employees Sorted by Department (A–Z) and Salary (High → Low).

```python
"""
    SELECT * FROM employee 
    ORDER BY department, salary DESC;
"""
```

Q5. List All Employees in the IT Department, Ordered by Hire Date (Newest First).

```python
"""
    SELECT * FROM employee 
    WHERE department = 'IT'
    ORDER BY hire_date DESC;
"""
```

Q9. What is the Difference Between a Primary Key and a Foreign Key?

- Primary Key: Uniquely identifies each record in a table, cannot be NULL or duplicate.

- Foreign Key: Refers to the primary key of another table, can have duplicates and NULL values (unless restricted).

Q10. What Are Constraints in SQL and Why Are They Used?

- Constraints are rules applied to table columns to enforce data integrity and accuracy.

    e.g.=> NOT NULL, UNIQUE, PRIMARY KEY, FOREIGN KEY, CHECK, DEFAULT
